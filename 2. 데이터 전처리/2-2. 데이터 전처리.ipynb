# cotents

```python
contents_1 = pd.read_csv('contents_01_clean.csv', encoding='cp949')
contents_2 = pd.read_csv('contents_02_clean.csv', encoding='cp949')
contents_3 = pd.read_csv('contents_03_clean.csv', encoding='cp949')
```

### 3개 파일 합치기

```python
contents = pd.concat([contents_1, contents_2, contents_3], axis=0)
contents
```

### 불필요한 컬럼 제거

```python
contents.drop(['Unnamed: 0','Unnamed: 0.1'], axis=1, inplace=True)
```

```python
contents.info()
```

```python
contents['log_dt'] == 2.02401000e+13
```

```python
contents.replace(2.02401000e+13, np.nan, inplace=True)
```

### datetime 생성

```python
contents.log_dt.unique()
```

```python
date = pd.to_datetime(contents.log_dt, format = "%Y%m%d%H%M%S")
```

```python
contents['datetime'] = date
contents['datetime']
```

```python
contents['day'] = contents['datetime'].apply(lambda x : x.day)
contents['day'].fillna(-1, inplace=True)
contents['day'] = contents['day'].astype(int)
```

```python
contents['hour'] = contents['datetime'].apply(lambda x : x.hour)
contents['hour'].fillna(-1, inplace=True)
contents['hour'] = contents['hour'].astype(int)
```

```python
contents.drop('log_dt', axis=1, inplace=True)
```

```python
contents
```

### 저장

```python
contents.to_csv('contents.csv', encoding='cp949')
```
