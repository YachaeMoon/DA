```python
contents = pd.read_csv('contents_20240301.csv', encoding='cp949')
contents
```

```python
contents_drop = contents[(contents['genre_of_ct_cl'].isin(['에로틱', '성인']))
                                                               | (contents['genre_of_ct_cl'] == '기타') 
                                                               & (contents['ct_cl'] == '기타') 
                                                               & (contents['series_nm'].isna())]
```

```python
contents_drop
```

```python
contents = contents.drop(contents_drop.index)
```

```python
contents
```

```python
contents.to_csv('contents_03_clean.csv', encoding='cp949')
```
