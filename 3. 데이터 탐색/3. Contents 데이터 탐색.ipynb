```python
# 프로그램 이름 정규화 함수
def normalize_program_name(name):
    # 회차 제거 (모범택시 2 10회 -> 모범택시 2)
    name = re.sub(r'\d+회', '', name)
    return name
```

### 4. 요일/공휴일별 관람횟수

```python
contents_4 = contents.copy()
contents_4
```

```python
# 'log_dt' 열에서 요일 추출 (0: 월요일, 1: 화요일, ..., 6: 일요일)
contents_4['day_of_week'] = contents_4['datetime'].dt.dayofweek

# 요일별 관람횟수 계산
view_count_by_day = contents_4.groupby('day_of_week').size()

# 요일 이름으로 변환
day_names = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']
view_count_by_day.index = [day_names[int(i)] for i in view_count_by_day.index]

# 결과 출력
print(view_count_by_day)
```

```python
from datetime import datetime

# 주어진 날짜가 휴일인지 아닌지 판단하는 함수
def is_holiday(date):
    # 주말인지 확인 (0: 월요일, 1: 화요일, ..., 6: 일요일)
    if date.weekday() >= 5:  # 토요일(5) 또는 일요일(6)
        return True
    # 공휴일이면서 주말이 아닌지 확인
    # 여기서는 간단하게 예시로 공휴일을 지정합니다.
    holidays = [datetime(2024, 1, 1),
                datetime(2024, 2, 9),
                datetime(2024, 2 ,12),
                datetime(2024, 3, 1)]
    if date in holidays:
        return True
    return False

# 'log_dt' 열에서 휴일 여부 추출
contents_4['is_holiday'] = contents_4['datetime'].apply(is_holiday)

# 휴일과 비휴일 관람횟수 계산
view_count_by_holiday = contents_4.groupby('is_holiday').size()

# 결과 출력
print(view_count_by_holiday)

```
